```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#load data

credit_card_data = pd.read_csv('/content/creditcard.csv')
```

Unbalanced label set, not enough fraudulent transaction

0 == Normal Transaction
1 == Fraudulent Transaction

```python
#split the data according to fraud and normal
normal = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]

print(normal.shape)
print(fraud.shape)
```

```python
#info on the data
fraud.Amount.describe()
```

```python
#info on data
normal.Amount.describe()
```

```python
#compare values for both transactions
credit_card_data.groupby("Class").mean()
```

UNDER-SAMPLING

To fix the issue of having significantly more normal transactions then fraudulent transactions will balance class distribution by reducing the number of samples in the majority class

```python
#match amount of entries as fraudulent
new_normal = normal.sample(n=492)
```

Concantinate both Data stes

```python
#create the new data set to be used
new_data_set = pd.concat([new_normal, fraud], axis=0)
```

```python
#verify size is as it should be
new_data_set.shape
```

```python
#check first 5 rows
new_data_set.head()
```

```python
#verify count
new_data_set['Class'].value_counts()
```

We see below the avergaes for the NORMAL Transactions are still the same

```python
#compare values for both transactions
credit_card_data.groupby("Class").mean()
```

SPLIT DATA:
  - Targets
  - Feautures

```python
X = new_data_set.drop(columns="Class", axis=1)
Y = new_data_set["Class"]
```

SPLIT X AND Y:
  - Training Data
  - Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify=Y, random_state=2)
```

MODEL TRAINING

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

MODEL EVALUATION

```python
#accuracy on training data
X_training_prediction = model.predict(X_train)
train_data_accuracy = accuracy_score(X_training_prediction, Y_train)
```

```python
print(train_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print(test_data_accuracy)
```
